```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm, tree
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import roc_curve


import xgboost
```

```python
classifiers = []
model1 = xgboost.XGBClassifier()
classifiers.append(model1)
model2 = svm.SVC()
classifiers.append(model2)
model3 = tree.DecisionTreeClassifier()
classifiers.append(model3)
model4 = RandomForestClassifier()
classifiers.append(model4)
model5 = GaussianNB()
classifiers.append(model5)

print(classifiers)
```

```python
model6 = NearestNeighbors(n_neighbors=1)
# classifiers.append(model6)
```

```python
for clf in classifiers:
    clf.fit(X_train, y_train)
    y_pred= clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy of {clf} is {acc}")
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix of %s is %s"%(clf, cm))
    fprValue, tprValue, thresholdsValue = roc_curve(y_test,y_pred)
    print(f"fprValue of {clf} is {fprValue}")
    print(f"tprValue of {clf} is {tprValue}")
    print(f"thresholdsValue of {clf} is {thresholdsValue}")


```
